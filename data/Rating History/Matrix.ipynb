```python
from datetime import datetime
from datetime import timedelta
import math
```

```python
cra_name = "ICRA"
rating_scale = 20
terminal_date = datetime(2018,3,31)
terminal_date
```

```python
cradf = pd.read_excel("Numeric_Rating "+cra_name+".xlsx")
cradf
```

```python
matrix = pd.DataFrame(columns = range(1,rating_scale+1))
for i in range (rating_scale+1):
    for j in range (rating_scale+1):
        matrix.loc[i,j] = int(0)
matrix
```

```python
cradf = cradf.drop(['Name of the Issuer'], axis=1)
sample = cradf.head()
sample
```

```python
def studycase (case = None):
    rating = case.pop(0)
#     print(rating)
    if (rating != 0):
        rating_date = case.pop(0)
        next_rating = case[0]
        if (math.isnan(next_rating)):
            next_rating_date = terminal_date
        else:
            next_rating_date = case[1]
#         print ("Rating->", rating, type(rating))
#         print ("Rating Date->", rating_date, type(rating_date))
#         print ("Next Rating->", next_rating, type(next_rating))
#         print ("Next Rating Date ->", next_rating_date, type(next_rating_date))
        yearsdifference = (next_rating_date - rating_date)/timedelta(days=365)
        matrix.loc[rating,rating] += int(yearsdifference)
        
#         print ("years difference -> ", yearsdifference, type(yearsdifference))
        if (not math.isnan(next_rating)):
            matrix.loc[rating,next_rating] += 1
            studycase(case)

# case= (cradf.loc[6287]).tolist()
# print(case)
# studycase(case)
```

```python
for i, item in cradf.iterrows():
    case = item.tolist()
    print (i)
    studycase(case)
```
