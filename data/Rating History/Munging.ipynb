```python
import pandas as pd
import os
print ("Imported Pandas and OS")
```

```python
cra_name = "SMERA"
mapping_style = "Numeric"
cra = pd.read_excel("Clean "+cra_name+".xlsm", index_col = "Sl.No.")
cra.head()
```

```python
map = pd.read_excel("Clean "+cra_name+" Mapping.xlsx", index_col = "Base")
map.head()
mapping = map.to_dict()[mapping_style]
print (mapping)
```

```python
mapping['FA+/CRISIL\rA1']
```

```python
for i, item in cra.iterrows():
    print (i, "-->")
    for j in list(item.index.values):
        try:
            psedo = mapping[item[j]]
        except:
            psedo = item[j]
        cra.loc[i,j] = psedo
    os.system('cls')
    
```
